## Code for Data Analysis 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df1 = pd.read_csv('World Population by country 2024.csv')
df1
```

```python
df2 = pd.read_csv('Cost_of_Living_Index_by_Country_2024.csv')
df2
```

## Merging the Datasets

```python
df_merged = df1.merge(df2, on='Country', how='inner')
df_merged
```

## Data Cleaning

```python
df_merged.drop(columns='Population 2023', inplace=True)
df_merged.head()
```

## Downloading Merged Dataset to CSV file

```python
df_merged.to_csv("/Users/danielkim/Desktop/merged_population_dataset.csv", index=False)
```

## Visualization #1

```python
df = df_merged

# Display basic information about the merged dataset
print("Dataset Overview:")
print(f"Total countries in analysis: {len(df)}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst few rows:")
print(df.head())
```

```python
# Convert population to millions for easier interpretation
df['Population_Millions'] = df['Population 2024'] / 1000000

# Create population categories for better analysis
def population_category(pop):
    if pop < 1:  # Less than 1 million
        return 'Very Small (<1M)'
    elif pop < 10:  # 1-10 million
        return 'Small (1-10M)'
    elif pop < 50:  # 10-50 million
        return 'Medium (10-50M)'
    elif pop < 100:  # 50-100 million
        return 'Large (50-100M)'
    else:  # Over 100 million
        return 'Very Large (>100M)'

df['Population_Category'] = df['Population_Millions'].apply(population_category)

# Display summary statistics
print("Summary Statistics:")
print(f"Population range: {df['Population_Millions'].min():.2f}M to {df['Population_Millions'].max():.2f}M")
print(f"Cost of Living Index range: {df['Cost of Living Index'].min()} to {df['Cost of Living Index'].max()}")
print(f"Average Cost of Living Index: {df['Cost of Living Index'].mean():.2f}")
```

```python
numeric_cols = df_merged.select_dtypes(include=[np.number]).columns
correlation_matrix = df_merged[numeric_cols].corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, 
            annot=True, 
            cmap='coolwarm', 
            fmt='.2f', 
            square=True,  
            cbar_kws={'shrink': 0.8})

plt.title('Correlation Heatmap - Population & Economic Indicators', 
          fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
